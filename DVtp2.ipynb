```python
import pandas as pd
```

#Part 1 â€“ Case Study: Cleaning and Preparing the Sales Dataset (Outlier Focus)

Task 1.1: Data Acquisition and Initial Review

```python
url = "https://raw.githubusercontent.com/PacktPublishing/Hands-on-Exploratory-Data-Analysis-with-Python/refs/heads/master/Chapter%204/sales.csv"
df_sales = pd.read_csv(url)

print("First 5 rows:")
display(df_sales.head())

print("Dataset Info:")
df_sales.info()
```

Task 1.2: Feature Engineering â€“ Calculating Total Price

```python
df_sales['TotalPrice'] = df_sales['Quantity'] * df_sales['UnitPrice']

print("Check new 'TotalPrice' column:")
display(df_sales.head())
```

Task 1.3: Outlier Detection and Filtering

```python
outliers = df_sales[df_sales['TotalPrice'] > 3_000_000]
print(f"Number of outlier transactions: {outliers.shape[0]}")
print(f"\nOutliers \n: {outliers.head(3)} \n\n")
df_cleaned_outliers = df_sales[df_sales['TotalPrice'] <= 3_000_000]

removed_rows = df_sales.shape[0] - df_cleaned_outliers.shape[0]
print(f"Rows before removing outlier filtering: {df_sales.shape[0]}")
print(f"Rows after removing outlier filtering: {df_cleaned_outliers.shape[0]}")
```

#Part 2 â€“ Pandas Wrangling and Variable Classification

Task 1.1: Data Acquisition and Initial Review

```python
# Count duplicated rows
duplicates_count = df_sales.duplicated().sum()
print(f"Number of duplicate rows: {duplicates_count}")

# Remove duplicates, keep the last occurrence
df_no_duplicates = df_sales.drop_duplicates(keep='last')

# Verify row count
print(f"Rows after removing duplicates: {df_no_duplicates.shape[0]}")
```

Task A.2: Handling Missing Data

```python
# Count missing values per column
print("Missing values per column:")
print(df_no_duplicates.isna().sum())

# Drop rows with any missing value (temporary)
df_dropna = df_no_duplicates.dropna()
rows_lost = df_no_duplicates.shape[0] - df_dropna.shape[0]
print(f"\nRows lost after dropping missing values: {rows_lost}")


```

Task A.3: Discretization and Binning

```python
labels = ['Low Quantity', 'Medium Quantity', 'High Quantity']

df_no_duplicates['Quantity_Bin'] = pd.cut(df_no_duplicates['Quantity'], bins=3, labels=labels)

# Analyze distribution
print("ðŸ”¹ Quantity Bin Distribution:")
print(df_no_duplicates['Quantity_Bin'].value_counts())
```
